```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
path = "/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv"
df = pd.read_csv(path)

df.isnull().sum()
```

```python
df.describe()


df.median(numeric_only=True)
```

```python
numeric_cols = ['Age', 'Fare']
df[numeric_cols].hist(bins=20, figsize=(10, 5))
plt.show()
```

```python
for col in numeric_cols:
    sb.boxplot(data=df, y=col)
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
sb.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']].dropna(), hue='Survived')
plt.show()


corr = df.corr(numeric_only=True)
sb.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
sb.barplot(data=df, x='Pclass', y='Survived')
plt.title("Survival Rate: Class")
plt.show()

sb.barplot(data=df, x='Sex', y='Survived')
plt.title("Survival Rate: Gender")
plt.show()
```
